```python
import os
import cv2
import matplotlib.pyplot as plt

# Define paths to the images
base_dir = './word/'
image_paths = [
    os.path.join(base_dir, 'image16.tif'),  # Original Image (RGB)
    os.path.join(base_dir, 'mask16_og.tif'),  # Original Mask (Grayscale)
    os.path.join(base_dir, 'resnet101_1e-05_Unet_128_noaug_image16.tif_pred_128.png'),  # 128
    os.path.join(base_dir, 'vgg16_1e-05_Unet_256_noaug_image16.tif_pred_256.png'),  # 256
    os.path.join(base_dir, 'vgg19_1e-05_Unet_512_noaug_image16.tif_pred_512.png'),  # 512
    os.path.join(base_dir, 'densenet201_1e-05_Unet_1024_noaug_image16.tif_pred_1024.png')  # 1024
]

# Define the labels
titles = ['Original Image', 'Original Mask', '128', '256', '512', '1024']

# Create a figure with a 3x2 grid layout
fig, axes = plt.subplots(3, 2, figsize=(12, 18))  # 3 rows, 2 columns, larger figure

# Load and plot each image with the corresponding title
for i, ax in enumerate(axes.flat):
    if i == 0:
        # Original image is RGB
        img = cv2.imread(image_paths[i])
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert from BGR to RGB for display
    else:
        # The rest are grayscale
        img = cv2.imread(image_paths[i], cv2.IMREAD_GRAYSCALE)
    
    # Display the image
    ax.imshow(img, cmap='gray' if i > 0 else None)
    
    # Add title below the image
    ax.set_title(titles[i], fontsize=12)
    
    # Remove axis ticks for cleaner display
    ax.set_xticks([])
    ax.set_yticks([])

# Adjust layout for better spacing
plt.tight_layout()

# Save the figure with high resolution (300 DPI)
output_path = os.path.join(base_dir, 'comparison_figure_high_res.png')
plt.savefig(output_path, dpi=300)

# Show the figure
plt.show()

print(f"Figure saved to: {output_path}")
```

```python
import os
import cv2
import matplotlib.pyplot as plt

# Define paths to the images
base_dir = './word/'
image_paths = [
    os.path.join(base_dir, 'image16.tif'),  # Original Image (RGB)
    os.path.join(base_dir, 'mask16_og.tif'),  # Original Mask (Grayscale)
    os.path.join(base_dir, 'resnet101_1e-05_Unet_128_noaug_image16.tif_pred_128.png'),  # 128
    os.path.join(base_dir, 'vgg16_1e-05_Unet_256_noaug_image16.tif_pred_256.png'),  # 256
    os.path.join(base_dir, 'vgg19_1e-05_Unet_512_noaug_image16.tif_pred_512.png'),  # 512
    os.path.join(base_dir, 'densenet201_1e-05_Unet_1024_noaug_image16.tif_pred_1024.png')  # 1024
]

# Define the labels
titles = ['Original Image', 'Original Mask', '128', '256', '512', '1024']

# Define the crop size
crop_size = 1024

# Function to crop an image from the top-left corner
def crop_image(image, crop_size):
    height, width = image.shape[:2]
    cropped_image = image[:min(crop_size, height), :min(crop_size, width)]
    return cropped_image

# Create a figure with a 3x2 grid layout
fig, axes = plt.subplots(3, 2, figsize=(12, 18))  # 3 rows, 2 columns, larger figure

# Load, crop, and plot each image with the corresponding title
for i, ax in enumerate(axes.flat):
    if i == 0:
        # Original image is RGB
        img = cv2.imread(image_paths[i])
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert from BGR to RGB for display
    else:
        # The rest are grayscale
        img = cv2.imread(image_paths[i], cv2.IMREAD_GRAYSCALE)
    
    # Crop the image to the top-left 2000x2000 section
    img_cropped = crop_image(img, crop_size)
    
    # Display the cropped image
    ax.imshow(img_cropped, cmap='gray' if i > 0 else None)
    
    # Add title below the image
    ax.set_title(titles[i], fontsize=12)
    
    # Remove axis ticks for cleaner display
    ax.set_xticks([])
    ax.set_yticks([])

# Adjust layout for better spacing
plt.tight_layout()

# Save the figure with high resolution (300 DPI)
output_path = os.path.join(base_dir, 'comparison_figure_cropped_high_res.png')
plt.savefig(output_path, dpi=300)

# Show the figure
plt.show()

print(f"Figure saved to: {output_path}")
```
