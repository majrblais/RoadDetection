```python
import os
from PIL import Image
import cv2
def convert_and_resize_tiff_to_png(input_folder, output_folder, size=(1024, 1024)):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    # Iterate over all files in the input folder
    for filename in os.listdir(input_folder):
        if filename.lower().endswith('.tif') or filename.lower().endswith('.tiff'):
            # Construct the full file path
            file_path = os.path.join(input_folder, filename)
            
            # Read the TIFF image
            img = cv2.imread(file_path, cv2.IMREAD_UNCHANGED)
            
            if img is not None:
                # Resize the image
                img_resized = cv2.resize(img, size, interpolation=cv2.INTER_AREA)
                
                # Create the output file path
                output_file_path = os.path.join(output_folder, os.path.splitext(filename)[0] + '.png')
                
                # Save the resized image as PNG
                cv2.imwrite(output_file_path, img_resized)
                
                print(f"Converted and resized {filename} to {output_file_path}")
            else:
                print(f"Failed to read {filename}")

# Define input and output folders
input_folder = './data/images2_b/'
output_folder = './data/images2_b_resized/'

# Convert and resize all TIFF images in the input folder
convert_and_resize_tiff_to_png(input_folder, output_folder)
```
